```python
#winifred.kwakye"azubiafrica.org
#Individual Project

import pandas as pd
#Creating a dictionary with required information:
data = {
  "product_id": [23, 96, 97, 15, 87],
  "Name": ["computer", "Python Workout", "Pandas Workout", "banana", "sandwich"],
  "wholesale_price":[500, 35, 35, 0.5, 3],
  "retail_price":[1000, 75, 75, 1, 5],
  "sales":[100, 1000, 500, 200, 300]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

```python
#Step2:Calculate the profit for each product:
df["net_revenue_per_product"] = (df["retail_price"] - df["wholesale_price"]) * df["sales"]
df
```

```python
#Indivual Project:The net revenue per product is being reduced by a 15, 20 and 25 percent taxes.
#Calculating the 15 percent sales tax:
Salex_tax_fifteen = (df["net_revenue_per_product"]*0.15)
df["net_revenue_per_product_reducedby15percent"] = (df["net_revenue_per_product"] - Salex_tax_fifteen)
df
```

```python
#Calculating the 20 percent sales tax:
Salex_tax_twenty = (df["net_revenue_per_product"]*0.20)
df["net_revenue_per_product_reducedby20percent"] = (df["net_revenue_per_product"] - Salex_tax_twenty )
df
```

```python
#Calculating the 25 percent sales tax:
Salex_tax_twentyfive = (df["net_revenue_per_product"]*0.25)
df["net_revenue_per_product_reducedby25percent"] = (df["net_revenue_per_product"] - Salex_tax_twentyfive )
df
```

```python
#Calculating the orignal net:
total_revenue = df["net_revenue_per_product"].sum()
print("The total net revenue recieved is",total_revenue)

#Calculating the net reduced by 15 percent sales tax:
netReducedBy15 = df["net_revenue_per_product_reducedby15percent"].sum()
print("The total net revenue reduced by 15 percent sales tax is",netReducedBy15)

#Calculating the net reduced by 20 percent sales tax:
netReducedBy20 = df["net_revenue_per_product_reducedby20percent"].sum()
print("The total net revenue reduced by 20 percent sales tax is",netReducedBy20)

#Calculating the net reduced by 25 percent sales tax:
netReducedBy25 = df["net_revenue_per_product_reducedby25percent"].sum()
print("The total net revenue reduced by 25 percent sales tax is",netReducedBy25)
```
